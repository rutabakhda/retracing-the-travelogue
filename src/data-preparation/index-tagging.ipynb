import os
import pandas as pd
import numpy as np


basepath = os.path.dirname(os.path.abspath('__file__'))
datapath = basepath + '/data/hugh-murray/index/'
readfile1 = 'index.csv'

df = pd.read_csv(datapath + readfile1,sep=',',encoding='ISO-8859-1',nrows=100)

df.head()

```python
from nltk.tag import StanfordNERTagger
from nltk.tokenize import word_tokenize
from nltk.tag import StanfordPOSTagger
import re

stanford_dir = "E:\python\standford-nlp\stanford-postagger-2018-10-16"
modelfile = stanford_dir+"\models\english-bidirectional-distsim.tagger"
jarfile=stanford_dir+"\stanford-postagger.jar"

def remove_special_characters(text):
    pattern = r'[^a-zA-z\s]'
    text = re.sub(pattern, '', text)
    return text

new_data = pd.DataFrame(columns=['Entity Name', 'Type', 'Page No','word','tag'])
for index,row in df.iterrows():
    if row[1]:
        text = remove_special_characters(row[1])
        print(text)
        new_row = {}
        new_row['Entity Name'] = row[0]
        new_row['Type'] = row[1]
        new_row['Page No'] = row[2]

        tagger=StanfordPOSTagger(model_filename=modelfile, path_to_jar=jarfile)
        tagged_text = tagger.tag(text.split())

        st = StanfordNERTagger('data/standford-nlp/stanford-ner-2014-06-16/classifiers/english.all.3class.distsim.crf.ser.gz', 'data/standford-nlp/stanford-ner-2014-06-16/stanford-ner.jar', encoding='utf-8')
        classified_text = st.tag(text.split())

        mylist = []
        for word,tag in tagged_text:
            if tag.startswith('VB') or tag.startswith('IN'):
                break
            else:

                mylist.append([word,tag])

        print(mylist)

        for word,tag in mylist:
            if tag.startswith('N'):
                final_noun_word = word

        print(final_noun_word)

        for word,tag in classified_text:
            if word == final_noun_word:
                if tag == 'O':
                    final_tag = 'PEOPLE'
                    print(final_tag)
                    break


        new_row['word'] = word
        new_row['tag'] = tag
        new_data.loc[len(new_data)] = new_row

new_data.to_csv(datapath + "index-processed.csv",sep=',',encoding='ISO-8859-1',index=False)


df.head()
```
